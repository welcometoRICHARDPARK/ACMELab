```python
title = soup.find('meta', attrs={'property':'og:title'})
title['content']
```

```python
test = soup.find_all('div', 'par')
```

```python
display(test[0])
display(test[1])
```

```python
news_description = []

for content in test:
    content = str(content)
    content = re.sub('<.+?>',"",content,0).strip()
    news_description.append(content)

# news_description = str(soup.find_all('div', 'par'))
# news_description = re.sub('<.+?>',"",news_description,0).strip()
```

```python
news = [news_description[0] + news_description[1]]
```

```python
news
```

```python
from konlpy.tag import Okt
```

```python
dir(Okt())
```

```python
okt = Okt()


# 뉴스 기사에서 명사만 추출
for i in news :
    nouns_list = okt.nouns(i)
    print(nouns_list)
```

```python
# 불용어 제거
print("[알림] 불용어를 제거합니다..")
nouns_without_stopwords = []
for i in nouns_list:
   if len(i) != 1:
       nouns_without_stopwords.append(i)
print(nouns_without_stopwords)

# 단어 빈도수 카운트
print("[알림] 단어 빈도수를 카운트합니다..")
from collections import Counter
count_result = Counter(nouns_without_stopwords)
print(count_result)

```
