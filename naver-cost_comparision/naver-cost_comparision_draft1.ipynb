```python
merchanidse = '삼다수'
keyword = '2l'
raw_query = merchanidse+keyword
query = quote(raw_query)

url = f'https://search.shopping.naver.com/search/all?frm=NVSHMDL&iq=%EC%82%BC%EB%8B%A4%EC%88%982l&origQuery=%EC%82%BC%EB%8B%A4%EC%88%98&pagingIndex=1&pagingSize=40&productSet=model&query={query}&sort=rel&timestamp=&viewType=list'
```

```python
response = req.urlopen(url)
soup = BeautifulSoup(response,'html.parser')
```

```python
search_results = soup.select('.basicList_title__3P9Q7 > .basicList_link__1MaTN')
search_results_list = []

for idx in range(len(search_results)) :
    search_results_dict = {}
    search_results_dict['title'] = soup.select('.basicList_title__3P9Q7 > .basicList_link__1MaTN')[idx]['title']
    search_results_dict['link'] = soup.select('.basicList_title__3P9Q7 > .basicList_link__1MaTN')[idx]['href']
    search_results_list.append(search_results_dict)
```

```python
(search_results_list[0]['title'])
```

```python
for idx in tqdm(range(len(search_results_list))):
    merchanidse_name = search_results_list[idx]['title']

    
    ################################## vendor_name, membership_name  ##################################
    child_url = search_results_list[idx]['link']
    child_response = req.urlopen(child_url)
    child_soup = BeautifulSoup(child_response, 'html.parser')

    vendor_membership_name = child_soup.select('.productList_mall_link__1-Q4X')
    
    for idx, name in (enumerate(vendor_membership_name)):
        result['merchandise_name'].append(merchanidse_name)
        result['vendor_membership_name'].append(name.text)
 
        

        ################################## 가격  ##################################
        chromeDriver_dir = '/Users/bagdaehui/Desktop/chromedriver'
        options = webdriver.ChromeOptions()
        options.add_argument('headless')
        driver = webdriver.Chrome(chromeDriver_dir, options=options)
        driver.get(child_url)

        ######## 가격 - 배송비 미 포함 ########
        price_without_delivery_fee = driver.find_element(By.XPATH,
            f'//*[@id="section_price"]/div[3]/ul/li[{idx+1}]/div/div[3]/a/span'
        ) 
        result['price_without_delivery_fee'].append(price_without_delivery_fee.text)

        

        ################################## 배송비 ##################################
        delivery_fee = driver.find_element(By.XPATH,
            f'//*[@id="section_price"]/div[3]/ul/li[{idx+1}]/div/div[3]/div[2]'
        )
        if delivery_fee.text == '':
            delivery_fee = driver.find_element(By.XPATH,
                f'//*[@id="section_price"]/div[3]/ul/li[{idx+1}]/div/div[3]/div[1]'
            )
        result['delivery_fee'].append(delivery_fee.text)
        
        
        
```
